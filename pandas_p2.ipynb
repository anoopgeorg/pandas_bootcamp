```python
# dataframe selection based on row index
df[0:5]
```

```python
# Row selection using iloc - Index location
# Always needs the inbuilt index for search and display
df.iloc[0:5]
```

```python
# iloc cannot handle search only with inbuilt indecies for both column and row
# here ['PassengerId','Survived'] are named index and not built-in index
df.iloc[0:5,['PassengerId','Survived']]
```

```python
# iloc cannot handle search only with inbuilt indecies for both column and row
# here instead of ['PassengerId','Survived'] we give [0,1] as it is the in built index
df.iloc[0:5,[0,1]]
```

```python
# Row selection using iloc - Location
# Note the higher bound index gets included unlike other search operations
df.loc[0:5]
```

```python
# loc can use both index for row selection and named-index search for column selection
df.loc[0:5,['PassengerId','Survived']]
```
